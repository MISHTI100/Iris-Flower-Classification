```python
from sklearn.impute import SimpleImputer
imputer_mean = SimpleImputer(missing_values = np.nan, strategy = 'mean')
data[['Selling_Price','Present_Price']] = imputer_mean.fit_transform(data[['Selling_Price','Present_Price']]).astype('object')
```

```python
data.head(10)
```

```python
data.info()
```

```python
df = data.drop(columns = ['Car_Name','Year','Fuel_Type','Selling_type','Transmission','Owner'])
```

```python
df
```

```python
y = df.iloc[:,0].values
x = df.drop(columns = df.columns[0]).values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```
